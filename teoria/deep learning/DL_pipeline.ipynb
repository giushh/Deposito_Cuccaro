# creazione di un modello funzionale 

100 dimensione dataset in input
64 neuroni nei layer intermedi
10 layer finali

ogni neurone ha pesi e bias
quindi 
100 * 64 = 6400, ogni neurone gestisce 100 dimensioni
in più ci aggiungo 64 bias (per ogni neurone)
poi 64 neuroni mandano info ai 10 neuroni success -> +10 ecc

creazione modello

```python
from keras.layers import Input, Dense
from keras.models import Model

inputs = Input(shape=(100,))
x = Dense(64, activation='relu')(inputs)
outputs = Dense(10, activation='softmax')(x)
model = Model(inputs=inputs, outputs=outputs)
```

vediamo il totale dei parametri

```python
model.summary()
```

## Compilazione del Modello

Prima di addestrare il modello, è necessario compilare il modello specificando l'ottimizzatore, la funzione di perdita e le metriche.

in questo caso utilizzeremo adam, che è l'algoritmo che aggiorna i pesi

la categorical crosstentropy calcola l'errore classe trovata dalla rete valutando anche la confidenza della rete sull'errore

se la rete sbaglia ed è molto sicura che ha sbaglaito, viene punita aggiustando molto i pesi, rispetto a quando è insicura (aggiusta meno)

serve alla back propagation per aggiornare i pesi

e ha come metrica l'accurancy

ciclo

prova a caso -> aggiorna i pesi per massimizzare la metrica -> riprova

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
